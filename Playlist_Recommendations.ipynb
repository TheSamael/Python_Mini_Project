```python
import random,string
songids=['SONGID'+str(i).zfill(2) for i in range(1,21)]
userid=[''.join('USER')+str(random.randint(10,99)).zfill(2) for i in range(100)]
playlist={i:j for i,j in zip(userid,[random.sample(songids,random.randint(4,7)) for i in range(len(userid))])}
playlist
```

#### 1. Pass in a list the songs currently listed by a new user. This can have entries like SONGID01,SONGID02..

```python
#Create a list variable 'myplist' containing song IDs 'SONGID01', 'SONGID06', 'SONGID16'.
myplist=['SONGID01','SONGID06','SONGID16']
```

#### 2. Find a playlist in the dictionary, where ALL of these songs are present (there can be more than 1 such playlist)

```python
#Empty list to store the matching playlist
matching_playlists = []
#Iterate over playlist, checking if myplist is a subset of songs for each user
for user, songs in playlist.items():
    #Check if the set of songs in "myplist" is a subset of the set of "songs" for the current iteration
    if set(myplist).issubset(set(songs)):
        #If true, append songs to matching_playlists list
        matching_playlists.append(songs)
```

#### 3. From this playlist(s), pick the remaining songs and recommend to the new user (refer below examples for more details)

```python
#Empty list to store recommended songs
recommended_songs = []
#Iterate over the matching playlists list, where each iteration assigns the current set of songs to the variable "songs"
for songs in matching_playlists:
    #Iterate over the current set of songs to check if the song is not already in myplist
    for song in songs:
        if song not in myplist:
            #If the song is not in myplist, append the current song to the "recommended_songs" list
            recommended_songs.append(song)
#Print the recommended_songs list
print(recommended_songs)
```
